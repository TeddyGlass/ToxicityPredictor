```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
```

```python
df = pd.read_csv('../../processed/230301_iFAERS_VlcanoMS_Significant_AplusB100_SMILES_Filled.csv')
smiles = df.SMILES.tolist()
```

```python
mols, dfs = [], []
for i, s in enumerate(smiles):
    mol = Chem.MolFromSmiles(s)
    if mol is None:
        print(i, df.iloc[i,0], s)
    else:
        atoms = [a.GetSymbol() for a in mol.GetAtoms()]
        if 'C' in atoms:
            mols.append(mol)
            dfs.append(pd.DataFrame(df.iloc[i,:]).T)
        else:
            print(atoms)
```

```python
df_re = pd.concat(dfs)
df_re.to_csv(
    '../../processed/230306_iFAERS_VlcanoMS_Significant_AplusB100_SMILES_Filled_1173.csv',
    index=False
    )
df_re
```

```python
writer = Chem.SDWriter('../ToxicityPredictor/data/20230306_iFAERS1173.sdf')
for mol in mols:
    writer.write(mol)
```
